```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")
```

```python
diabetes = pd.read_csv('diabetes.csv')
diabetes               
```

```python
sns.scatterplot(x=diabetes['BMI'], y=diabetes['Glucose'])

plt.show()
```

```python
sns.relplot(data=diabetes, x="BMI", y="Glucose", hue="Pregnancies")
```

```python
sns.relplot(data=diabetes, x="BMI", y= "Glucose", hue="Pregnancies", size="Pregnancies", sizes=(10,50))
plt.xlim(18,37)
plt.ylim(50,140)

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set(style="whitegrid")
```

```python
Salary_data= pd.read_csv("Salary Data.csv")
Salary_data.head()
```

```python
plt.figure()
sns.scatterplot(x="Age", y="Salary", data=Salary_data)
plt.title('Default Scatter Plot: Age vs Salary')
plt.show()
                
```

```python
plt.figure()
sns.scatterplot(x="Age", y="Salary", hue="Education Level", style="Gender", data=Salary_data)
plt.title('Scatter Plot: Age vs Salary by Education Level and Gender')
plt.show()
```

```python
education_colors= {"Bachelor's": "green", "Master's": "blue", "PhD": "red"}
plt.figure()
sns.scatterplot(x="Age", y="Salary", hue="Education Level", style="Gender", data=Salary_data, palette=education_colors)
plt.title('Scatter Plot: Age vs Salary by Education Level and Gender')
plt.show()
```

```python
Gender_markers = {"Female": "o", "Male": "s"}
education_colors= {"Bachelor's": "green", "Master's": "blue", "PhD": "red"}
plt.figure()
sns.scatterplot(x="Age", y="Salary", hue="Education Level", style="Gender", data=Salary_data, palette=education_colors)
plt.title('Scatter Plot: Age vs Salary by Education Level and Gender')
plt.show()
```

```python
Gender_markers = {"Female": "o", "Male": "s"}
education_colors= {"Bachelor's": "green", "Master's": "blue", "PhD": "red"}

plt.figure(figsize=(12,8))
sns.set_style("ticks")
sns.scatterplot(x="Age", y="Salary", hue="Education Level", style="Gender", data=Salary_data, palette=education_colors)
plt.title('Scatter Plot: Age vs Salary by Education Level and Gender')
plt.show()
```

```python
Gender_markers = {"Female": "o", "Male": "s"}
education_colors= {"Bachelor's": "green", "Master's": "blue", "PhD": "red"}

plt.figure(figsize=(12,8))
sns.set_style("ticks")
sns.scatterplot(x="Age", y="Salary", hue="Education Level", style="Gender", data=Salary_data, palette=education_colors)

plt.legend(ncol=2, frameon=False, loc="lower right")
plt.title('Scatter Plot: Age vs Salary by Education Level and Gender')
plt.show()
```

```python
gender_colors= {"Female": "red", "Male": "blue"}

plt.figure(figsize=(10,8))
sns.set_style("ticks")
Bachelors=sns.scatterplot(x="Age", y="Salary", hue="Gender", data=Salary_data[Salary_data["Education Level"] == "Bachelor's"], palette=gender_colors)

Bachelors.set_xlim(25,50)
Bachelors.set_ylim(50000, 2500000)
plt.legend(ncol=2, frameon=False, loc="upper left")
plt.title('Scatter Plot: Age vs Salary by Education Level and Gender')
plt.show()
```
